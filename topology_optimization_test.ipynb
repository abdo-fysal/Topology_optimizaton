```python
from tkinter import Tcl
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.python.keras.models import Model,load_model
```

```python
def get_data(data_path):
    x_img=[]
    l1=[]
    l2=[]
    l3=[]
    l4=[]
    l5=[]
    files=Tcl().call('lsort', '-dict', os.listdir(data_path))
    i=0
    while i < len(files):

        print(i)





        for j in range(5):

            normalizedImg = np.zeros((256, 256),dtype='float32')

            img = cv2.imread(data_path+"//"+files[i+j],0)
            print(data_path+"//"+files[i+j])

            img = cv2.normalize(img,  normalizedImg, 0, 255, cv2.NORM_MINMAX)
            plt.imshow(img,cmap='gray')
            plt.show()
            img = cv2.resize(img,(256,256))

            img=np.reshape(img,(256,256,1))

            if j==0:
                l1.append(img)

            if j==1:
                l2.append(img)

            if j==2:
                l3.append(img)

            if j==3:
                l4.append(img)






            if j==4:
                normalizedImg = np.zeros((256, 256),dtype='float32')

                img = cv2.imread(data_path+"//"+files[i+j],0)
                print(data_path+"//"+files[i+j])
                plt.imshow(img,cmap='gray')
                plt.show()


                img = cv2.normalize(img,  normalizedImg, 0, 255, cv2.NORM_MINMAX)
                img = cv2.resize(img,(256,256))

                img=np.reshape(img,(256,256,1))
                l5.append(img)



                
                i+=5

    

    x_img=np.array([l1,l2,l3,l4,l5],dtype='float32')
    x_img=np.swapaxes(x_img,0,4)
    x_img=x_img.reshape((x_img.shape[0],256,256,5))
    
    return x_img

```

```python
input_img=get_data('C:\\Users\\abdo\\Desktop\\test')
```

```python
model = load_model ('C:\\Users\\abdo\\Desktop\\model_013680.h5')
generated_image=model(input_img)
```

```python
plt.imshow(generated_image[0,:,:,0],cmap='gray')
plt.show()
plt.savefig('optimized_image')
plt.close()
```
